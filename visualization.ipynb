```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load Data

## Purple Air

```python
pa_sensor_id_list = ["155565", "163173"]
df_pa1 = pd.read_csv("data/pa_10162023_10232023_155565(30MinMean).csv")
df_pa2 = pd.read_csv("data/pa_10162023_10232023_163173(30MinMean).csv")
df_pa1.head()
```

## SJV Air

```python
sj_sensor_id_list = ["9tYuybX_ROqhTm5m6je79Q", "CCA4NsviS-K4reZPsMUTYQ"]
df_sj1 = pd.read_csv("data/sj_10162023_10232023_9tYuybX_ROqhTm5m6je79Q(30MinMean).csv")
df_sj2 = pd.read_csv("data/sj_10162023_10232023_CCA4NsviS-K4reZPsMUTYQ(30MinMean).csv")
df_sj1.head()
```

## Air Gradient

```python
ag_sensor_id_list = ["ucm"]
df_ag1 = pd.read_csv("data/ag_10162023_10232023_ucm(30MinMean).csv")
df_ag1.head()
```

## Unify Column Name

```python
df_pa1 = df_pa1.rename(columns={"pm2.5_alt": "pm25"})
df_pa2 = df_pa2.rename(columns={"pm2.5_alt": "pm25"})
df_pa1.head()
```

# Visualize Data

## Purple Air v.s. Air Gradient

```python
fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(df_pa1["pm25"], label="Purple Air 1")
ax.plot(df_pa2["pm25"], label="Purple Air 2")
ax.plot(df_ag1["pm25"], label="Air Gradient")

plt.legend(fontsize=20)
plt.title("PM2.5 Purple Air v.s. Air Gradient", fontsize=30)
plt.show()
```

## Purple Air v.s. SJV Air

```python
fig, axes = plt.subplots(2, 1, figsize=(20, 20))
ax1 = axes[0]
ax2 = axes[1]

ax1.plot(df_pa1["pm25"], label="Purple Air 1")
ax1.plot(df_sj1["pm25"], label="SJV Air 1")
ax2.plot(df_pa2["pm25"], label="Purple Air 2")
ax2.plot(df_sj2["pm25"], label="SJV Air 2")

ax1.legend(fontsize=20)
ax2.legend(fontsize=20)
ax1.set_title("PM2.5 Purple Air v.s. SJV Air (Sensor 1)", fontsize=30)
ax2.set_title("PM2.5 Purple Air v.s. SJV Air (Sensor 2)", fontsize=30)
plt.show()
```

## SJV Air v.s. Air Gradient

```python
fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(df_sj1["pm25"], label="SJV Air 1")
ax.plot(df_sj2["pm25"], label="SJV Air 2")
ax.plot(df_ag1["pm25"], label="Air Gradient")

plt.legend(fontsize=20)
plt.title("PM2.5 SJV Air v.s. Air Gradient", fontsize=30)
plt.show()
```
